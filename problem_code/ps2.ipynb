## Problem 5 - Section 2.6 - Question 25

```python
import numpy as np
from scipy.linalg import toeplitz, lu
```

```python
# Create Toeplitz matrix
K = toeplitz([2, -1, 0, 0, 0, 0])

# LU decomposition (SciPy returns P, L, U)
P, L, U = lu(K) # type: ignore

# Inverse of L
L_inv = np.linalg.inv(L)

print(L_inv)
```
