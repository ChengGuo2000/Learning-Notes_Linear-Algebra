```python
# Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up
A = np.array([[0.8, 0.3], 
              [0.2, 0.7]])
u = np.array([[1],
              [0]])
v = np.array([[0],
              [1]])
```

```python
# Calculate for u
x = u.copy()
k = np.arange(0, 8)
while x.shape[1] <= 7:
    u = A @ u
    x = np.hstack((x, u))

# Plot for u
plt.plot(k, x[0, :], label="x1")
plt.plot(k, x[1, :], label="x2")
plt.xlabel("k")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate for v
x = v.copy()
k = np.arange(0, 8)
for j in range(7):
    v = A @ v # type: ignore
    x = np.hstack((x, v))

# Plot for v
plt.plot(k, x[0, :], label="x1")
plt.plot(k, x[1, :], label="x2")
plt.xlabel("k")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
